```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#1. load the data as an N * p array
iris_array = np.loadtxt("/Users/fanyingtang/Desktop/IRIS/iris.data.txt", delimiter=",", usecols =range(0,4) )
print("The iris.data has %s samples, with %s attributes per sample"% (iris_array.shape[0], iris_array.shape[1]))

#2. generate a vector containing the samples' labels 
# import data as dataframe and select the column with the labels
df = pd.read_csv("/Users/fanyingtang/Desktop/IRIS/iris.data.txt", header = None)
irislabels = df.iloc[:, 4]

# change the dataframe into an array
irislabels = irislabels.values

# count the number of samples in each species after changing the array into a list
iris_list = np.array(irislabels).tolist()
uniq_species = set(iris_list)
print(uniq_species)
for item in uniq_species:
    print(str(item) + ":" +str(iris_list.count(item)) + " samples")

```

```python
# 3. Generate the 2D scatterplots
sepal_length = df.iloc[:,0] 
sepal_width = df.iloc[:,1] 
petal_length = df.iloc[:,2] 
petal_width = df.iloc[:,3] 
iris_features = [sepal_length, sepal_width, petal_length, petal_width]
iris_attributes = ["sepal_length", "sepal_width", "petal_length", "petal_width"]

colors = []
for r in range(50):
    colors.append("r") # for "Iris-setosa"
for g in range(50):
    colors.append("g") # for "Iris-versicolor"
for b in range(50):
    colors.append("b") # for "Iris-virginica"


#subplot(nrows, ncols, plot_number)
fig = plt.figure()
fig.suptitle('Iris Data (red=setosa, green=versicolor, blue=virginica)', fontsize=30, fontweight='bold')
fig.set_size_inches(16,16)


for i in range(4):
    for j in range(4):
        ax = fig.add_subplot(4, 4, (j+1+ i*4))
        plt.xlabel(iris_attributes[j])
        plt.ylabel(iris_attributes[i])
        ax.scatter(iris_features[j], iris_features[i], c = colors)
plt.show()
```
